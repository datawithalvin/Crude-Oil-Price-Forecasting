```python
# import standard libraries
import numpy as np
import pandas as pd
import glob
import math
import time
import os
import datetime
from datetime import date, timedelta

import warnings
warnings.filterwarnings("ignore")


## import statistical library
from scipy.stats import spearmanr
import scipy.stats as stats


# import viz libraries
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline


# import model libraries
from keras.models import load_model
from keras.utils.vis_utils import plot_model
from sklearn.preprocessing import MinMaxScaler
```

```python
# forecasting function crude oil

def forecast_crude_oil():
        # load latest dataset
        latest_data = "../historical-datasets/Crude OIL WTI Futures Historical Data.csv"


        # load model
        use_model = load_model("../models/crude-oil-model.h5")


        # load dataset
        latest_df = pd.read_csv(latest_data, parse_dates=["Date"], index_col=["Date"]).sort_index(ascending=True)


        # do some data preprocessing
        df = latest_df.reset_index()["Price"]
        scaler = MinMaxScaler(feature_range=(0,1))
        df = scaler.fit_transform(np.array(df).reshape(-1,1))

        train_size = int(len(df)*0.65)
        test_size = len(df) - train_size
        train_data, test_data = df[0:train_size,:], df[train_size:len(df),:1]

        x_input_test = test_data[-30:].reshape(1,-1)
        temp_input = list(x_input_test)
        temp_input = temp_input[0].tolist()

        # predict future concentration
        # predict for 3 week (21 days)
        future_list = []
        n_steps = 30
        i = 0

        while(i<14):
                if (len(temp_input)>30):
                        x_input_test=np.array(temp_input[1:])
                        print("{} day input {}".format(i,x_input_test))
                        x_input_test = x_input_test.reshape(1,-1)
                        x_input_test = x_input_test.reshape((1,n_steps,1))
                        y_hat = use_model.predict(x_input_test, verbose=0)
                        print("{} day output {}".format(i, y_hat))
                        temp_input.extend(y_hat[0].tolist())
                        temp_input = temp_input[1:]
                        future_list.extend(y_hat.tolist())
                        i = i + 1
                else:
                        x_input_test = x_input_test.reshape((1,n_steps,1))
                        y_hat = use_model.predict(x_input_test, verbose=0)
                        print(y_hat[0])
                        temp_input.extend(y_hat[0].tolist())
                        print(len(temp_input))
                        future_list.extend(y_hat.tolist())
                        i = i + 1


        # build future dataframe
        today_date = date.today()
        days = pd.date_range(start=today_date, periods=14)

        future_list = scaler.inverse_transform(future_list)
        future_df = pd.DataFrame(future_list)
        future_df.insert(0, "Date", days)
        future_df = future_df.rename(columns={0:"PredictPrice"})
        future_df["PredictPrice"] = future_df["PredictPrice"].round(2)
        future_df.sort_index(inplace=True)
        future_df = future_df.set_index(pd.DatetimeIndex(future_df["Date"]))


        # build historical + future dataframe
        future_df.drop(columns="Date", inplace=True)
        future_df.columns = ["Price"]

        # copy hist dataset
        newest_df = latest_df[["Price"]][:-1]

        # append the selected rows to the new dataframe
        new_rows = future_df[["Price"]]
        newest_df = newest_df.append(new_rows)


        # calculate its standard deviation
        # Resample the dataframe to weekly and calculate the standard deviation for the time series values
        yesterday = today_date - timedelta(days=1)
        yesterday = yesterday.strftime("%Y-%m-%d")
        weekly_data = newest_df[['Price']][:yesterday] 
        weekly_data.columns = ["stdev_weekly"]
        weekly_data = weekly_data.resample('7D').std()

        combined = newest_df.join(weekly_data, how="outer")


        # calculate its confidence low and high
        # Extract the time series values and uncertainty values (e.g. standard deviation) from the daily dataframe
        values = combined['Price']
        uncertainty = combined['stdev_weekly']

        # Calculate the 95% confidence interval for the time series values in the daily dataframe
        confidence_interval = stats.norm.interval(0.95, loc=values, scale=uncertainty)

        # Add the confidence interval values from the resampled daily dataframe to the weekly dataframe as new columns
        combined['confidence_low'] = confidence_interval[0]
        combined['confidence_high'] = confidence_interval[1]


        # take 120 days back date, fillna with interpolate
        threemonths = combined[-120:]
        threemonths = threemonths.interpolate(method='time',limit_direction='both')

        return newest_df, threemonths
```

```python
# call function on new variable
df, for_figures = forecast_crude_oil()
```

Create new column that contains price from previous day and previous week. Calculate its difference by subtract between previous and current price.

```python
df['PrevDay'] = df['Price'].shift(1) # new col, shift data by one day
df['PrevWeek'] = df['Price'].shift(7) # new col, shift data by one week

# subtract shifted values from original values
df["PriceDiffDay"] = df['Price'] - df['PrevDay']
df["PriceDiffWeek"] = df['Price'] - df['PrevWeek']

# display last five rows
df.tail()
```

Split the dataset between historical dataset and forecast dataset.

```python
today_date = date.today()
yesterday = today_date - timedelta(days=1)
yesterday = yesterday.strftime("%Y-%m-%d")
today_date = today_date.strftime("%Y-%m-%d")
```

```python
hist_df = df[:yesterday]
future_df = df[today_date:]

# display hist_df
hist_df.tail(10)
```

```python
# display difference distribution (day)
ax = sns.distplot(hist_df["PriceDiffDay"], hist=True, kde=True, 
             bins=int(150/5), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 3})
```

```python
# display difference distribution (week)
ax = sns.distplot(hist_df["PriceDiffWeek"], hist=True, kde=True, 
             bins=int(150/5), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 3})
```

```python
# estimate parameters of log-normal distribution that best fits data
mu, sigma = np.mean(np.log(hist_df["PrevWeek"])), np.std(np.log(hist_df["PrevWeek"]))

# generate random samples from log-normal distribution
samples = np.random.lognormal(mu, sigma, size=1000)

# create distplot of log-normal distribution with title

ax = sns.distplot(samples, hist=True, kde=True,
                bins=int(150/5), color="darkblue",
                hist_kws={'edgecolor':'black'},
                kde_kws={'linewidth': 3})
                
plt.title("Weekly Log-Normal Distribution")
```

```python
# calculate the mean and standard deviation from the historical dataframe as a reference
mean_day = hist_df['PriceDiffDay'].mean()
std_day = hist_df['PriceDiffDay'].std()

mean_week = hist_df['PriceDiffWeek'].mean()
std_week = hist_df['PriceDiffWeek'].std()

# calculate price difference on the future_df
future_df['ZScoreDay'] = future_df['PriceDiffDay'].apply(lambda x: (x - mean_day)/std_day)
future_df['ZScoreWeek'] = future_df['PriceDiffWeek'].apply(lambda x: (x - mean_week)/std_week)

# display future_df
future_df
```

Use weekly Z-score to determine if the forecasted values it's expected on the distribution or not.

```python
# take the today forecast price and determine where the z-score fall
latest_zscore = future_df["ZScoreWeek"].loc[today_date]
latest_zscore
```

```python
stdev = 0
direction = []


if latest_zscore < 0:
        if -2.0 > latest_zscore:
                stdev += 3
        elif -1.0 > latest_zscore <= -2.0:
                stdev += 2
        elif -1.0 >= latest_zscore < 0:
                stdev += 1
        direct = "left"
        direction.append(direct)
elif latest_zscore > 0:
        if latest_zscore > 3:
                stdev += 3
        elif 2 > latest_zscore <= 3:
                stdev += 2
        elif 0 > latest_zscore <= 1:
                stdev += 1
        direct = "right"
        direction.append(direct)


print(stdev, "stdev to", direction[0])
```
